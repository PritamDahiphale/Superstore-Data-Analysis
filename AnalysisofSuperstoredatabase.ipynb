Hypothesis Testing

```python
import scipy.stats as stats
import pandas as pd
```

```python
df=pd.read_csv("SampleSuperstore.csv")

# Extract data for East and West regions
east_sales = df[df['Region'] == 'South']['Sales']
west_sales = df[df['Region'] == 'West']['Sales']

# Perform independent samples t-test
t_statistic, p_value = stats.ttest_ind(east_sales, west_sales)

# Print the results
print("T-statistic:", t_statistic)
print("P-value:", p_value)

# Interpret the results
alpha = 0.05  # Set the significance level
if p_value < alpha:
    print("Reject the null hypothesis. There is a significant difference in mean sales between the East and West regions.")
else:
    print("Fail to reject the null hypothesis. There is no significant difference in mean sales between the East and West regions.")
```

Discriptive analysis

```python
# Display basic statistics of numerical columns
print(df.describe())

# Calculate and display the total sales
total_sales = df['Sales'].sum()
print("Total Sales:", total_sales)

# Calculate and display the average sales
average_sales = df['Sales'].mean()
print("Average Sales:", average_sales)

# Calculate and display the maximum and minimum sales
max_sales = df['Sales'].max()
min_sales = df['Sales'].min()
print("Maximum Sales:", max_sales)
print("Minimum Sales:", min_sales)

# Calculate and display the total profit
total_profit = df['Profit'].sum()
print("Total Profit:", total_profit)
```

Product Analysis - Total Sales and Profit by Category

```python
# Calculate total sales and profit by category
category_sales = df.groupby('Category')['Sales'].sum()
category_profit = df.groupby('Category')['Profit'].sum()

# Display the results
print("Total Sales by Category:")
print(category_sales)
print("\nTotal Profit by Category:")
print(category_profit)
```

Customer Analysis - Average Order Value by Customer Segment

```python
# Calculate average order value by customer segment
segment_order_value = df.groupby('Segment')['Sales'].mean()

# Display the results
print("Average Order Value by Customer Segment:")
print(segment_order_value)
```

Region Analysis - Average Sales and Profit by Region

```python
# Calculate average sales and profit by region
region_sales = df.groupby('Region')['Sales'].mean()
region_profit = df.groupby('Region')['Profit'].mean()

# Display the results
print("Average Sales by Region:")
print(region_sales)
print("\nAverage Profit by Region:")
print(region_profit)
```

Time Series Analysis - Sales Trends Over Time

```python
import matplotlib.pyplot as plt
# Group by category and calculate total sales for each category
category_sales = df.groupby('Category')['Sales'].sum()

# Plot the sales trends
plt.figure(figsize=(10, 6))
plt.bar(category_sales.index, category_sales.values)
plt.xlabel('Category')
plt.ylabel('Sales')
plt.title('Sales Trends by Category')
plt.xticks(rotation=45)
plt.show()
```
